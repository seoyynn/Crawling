### 멜론차트 Top100 음원데이터 수집
- 가수명, 곡명 데이터 수집
- 가수명, 곡명이라는 키값으로 설정해서 딕셔너리로 정리
- 만들어진 데이터는 pandas를 이용해서 DataFrame으로 생성
- DataFrame을 엑셀파일로 저장

```python
# 환경세팅
# html요청
# 선택자 접근
# header 설정
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
```

```python
# 멜론 > 멜론차트 > Top100
# url 가져오기
melon_url = 'https://www.melon.com/chart/index.htm'
res = req.get(melon_url, headers = header) # res = req.get(melon_url, header) --> headers = header 를 header라고 생략가능
```

```python
res
```

```python
# 파이썬객체로 변환
melon_html = bs(res.text, 'lxml')
melon_html
```

```python
# 페이지 내의 html 태그 및 속성 규칙 확인
# 곡명, 가수명 데이터 수집
# 100개 개수가 맞는지 길이 확인
song_cnt = 0
song = melon_html.select('.rank01 > span > a')

for i in range(100):
    song_txt = song[i].text
    song_cnt += 1
    print(song_txt)
```

```python
singer_cnt = 0
singer = melon_html.select('.rank02')

for i in range(100):
    singer_txt = singer[i].text
    singer_cnt += 1
    print(singer_txt)
```

```python
print(song_cnt, singer_cnt)
```

```python
print(len(song), len(singer))
```

```python
song = melon_html.select('.ellipsis.rank01 a')
singer = melon_html.select('.checkEllipsis')
print(len(song), len(singer)) # 개수확인 꼭 하기

# 텍스트로 변환해서 각각의 리스트에 저장하기
song_txt = []
singer_txt = []
rank_lst = []
for i in range(len(song)):
    rank_lst.append(i+1)
    song_txt.append(song[i].text)
    singer_txt.append(singer[i].text)
print(len(song_txt), len(singer_txt))
```

```python
song
```

### 데이터프레임 생성

```python
# 1. 딕셔너리 정리
melon_dict = {'순위' : rank_lst, '노래 제목' : song_txt, '가수 이름' :singer_txt}
# 2. pd.DataFrame()
melon_df = pd.DataFrame(melon_dict)
melon_df2 = melon_df.set_index('순위') # .set_index() --> index이름으로 사용하겠다
melon_df2
```

```python
melon_df2.index # 순위는 int형태(숫자)

# 인덱징(접근) 숙제
# 순위가 30 ~ 60순위에 있는 가수명 출력해보기
melon_df2.loc[30:60, '가수 이름']
```

```python
# 엑셀 저장 : .to_excel('경로 및 파일명.xlsx', index = 'True = index포함저장, False = index제외하고 저장')
# csv 저장 : df.to_csv('경로 및 파일명.csv', index = 'True or False')
melon_df.to_excel('멜론차트Top100.xlsx', index = False)
```
